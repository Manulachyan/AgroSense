```python
fert_df = pd.read_csv("data_core.csv")
yield_df = pd.read_csv("yield_df.csv")

yield_df = yield_df.drop(columns=["Unnamed: 0"], errors="ignore")
yield_df = yield_df.rename(columns={"hg/ha_yield": "Yield"})
```

```python
def simulate_sensor_data():

    return {
        "Temparature": random.randint(20,35),
        "Humidity": random.randint(40,80),
        "Moisture": random.randint(20,70),
        "Nitrogen": random.randint(10,60),
        "Potassium": random.randint(10,60),
        "Phosphorous": random.randint(10,60),
        "Soil Type": random.choice(["Sandy","Loamy","Black","Red","Clayey"]),
        "Crop Type": random.choice(["Wheat","Rice","Maize","Cotton","Sugarcane"])
    }
```

```python
def predict_fertilizer(sensor_data):

    X_fert = fert_df.drop("Fertilizer Name", axis=1)
    X_fert = pd.get_dummies(X_fert)

    fert_features = X_fert.columns

    df_input = pd.DataFrame([sensor_data])
    df_input = pd.get_dummies(df_input)

    df_input = df_input.reindex(columns=fert_features, fill_value=0)

    prediction = fert_model.predict(df_input)

    return prediction[0]
```

```python
def predict_yield(sensor_data):

    area = random.choice(yield_df["Area"].unique())
    crop = random.choice(yield_df["Item"].unique())

    yield_input = {
        "average_rain_fall_mm_per_year": random.randint(800,1500),
        "avg_temp": sensor_data["Temparature"],
        "Year": random.choice(yield_df["Year"].unique()),
        "Area": area,
        "Item": crop,
        "pesticides_tonnes": random.randint(100,500)
    }

    df_input = pd.DataFrame([yield_input])

    # One-hot encode
    df_input = pd.get_dummies(df_input)

    # Force the same feature order used during training
    df_input = df_input.reindex(columns=yield_feature_order, fill_value=0)

    prediction = yield_model.predict(df_input)

    return prediction[0]
```

```python
while True:

    sensor_values = simulate_sensor_data()

    fertilizer = predict_fertilizer(sensor_values)

    predicted_yield = predict_yield(sensor_values)
    print("\n------ Smart Agriculture System ------")

    print("\nTemperature:", sensor_values["Temparature"], "°C")
    print("Humidity:", sensor_values["Humidity"], "%")
    print("Soil Moisture:", sensor_values["Moisture"])

    print("\nRecommended Fertilizer:", fertilizer)

    print("Predicted Crop Yield:", round(predicted_yield/10000,2), "tonnes/hectare")

    time.sleep(5)
```
